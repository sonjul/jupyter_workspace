# KRX 주식 종목 수집(KOSPI, KODAQ, KONEX 한꺼번에 수집)
### 컬럼정리
- Code : 종목코드
- ISU_CD : ISIN 코드(금융상품 식벽 국제 표준 코드)
- Name: 종목명
- Market : KOSPI, KODAQ, KONEX 구분값
- Dept : 소속부
- Close : 수집일 기준 종가
- ChnageCode : 등락기호, 가장 최근거래일의 가격 변동을 표시하는 코드(0, 1, 2, 3, 4, 5) ???
- ChangeRatio : 전일대비 등략률
- Open : 시가
- High : 고가
- Low : 저가
- Volume : 거래량
- Amount : 거래대금
- Marcap : 시가총액(원)
- Stocks : 상장주식수
- MarketId : 시장기호 ?
