# boolean indexing
- 결과값이 True, False bool 형태로 반환되기 때문에 boolean indexing 이라고 부른다.
- boolean indexing 을 사용해 특정 조건식을 만족하는 데이터를 서브셋으로 가져온다.

```python
df[df["Name"] == "카카오"]
```
