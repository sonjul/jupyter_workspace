# 1.3 한국거래소 종목정보 불러오기

```python
stocks = fdr.StockListing('KRX')
```

```python
stock_name = '삼성전자'
# stocks['Name'] == stock_name

def get_krx_code_by_name(krx_name):
    if krx_name in stocks['Name'][0]:
        krx_item_code = stocks[stocks['Name'] == stock_name]['Code'][0]
        print(f"종목 있어요 {krx_item_code}")
        return krx_item_code 
    else:
        print("종목 없어요.")
        raise '해당 종목코드는 없어요.'
        
def get_krx_trade_data(krx_name, start_date = None, end_date = None):
    krx_item_code = get_krx_code_by_name(krx_name)
    df = fdr.DataReader(krx_item_code, start= start_date, end = end_date)
    return df

df = get_krx_trade_data(stock_name, '2023-08-01', '2023-09-10')
        
# stock_code = get_krx_code_by_name(stock_name)
# df = fdr.DataReader(stock_code, start = '2023')
# df
```

### mdcd오실레이터

```python
def macd_oscillator(df, short_n, long_n, signal_n):
    df = df[['Close']].copy()
    
    df['Short'] = df['Close'].ewm(span = short_n, adjust = False).mean()
    df['Long'] = df['Close'].ewm(span = long_n, adjust = False).mean()
    df['MACD'] = df['Short'] - df['Long']
    df['Signal'] = df['MACD'].ewm(span = signal_n, adjust = False).mean()
    df['MACD Oscillator'] = df['MACD'] - df['Signal']
    
    return df

macd_oscillator(df, short_n = 9, long_n = 26, signal_n = 13)
```

# DD &MDD

- pct_change() : 수익률을 구하는 함수
- comport() : 누적곱을 구하는 함수(누적수익률)

- 일별 수익률 % = (오늘의 종가 - 어제의 종가) / 어제의 종가 * 100

```python
dd_df = df[['Close']]
((dd_df['Close'] - dd_df['Close'].shift(1)) / dd_df['Close'].shift(1)).fillna(0)
```

```python
def return_df(df):
    return df['Close'].pct_change().fillna(0)

# 일별 수익률
df_return = return_df(df)
df_return
```

```python
# 누적곱
def cum_return_fn(df):
    return (1 + df).cumprod()

df_cum_reutrn = cum_return_fn(df_return)


# 누적수익률 = 누적곱 - 1
df_cum_reutrn - 1
```

```python
df['Return'] = return_df(df)
df['CumReturn'] = cum_return_fn(df['Return'])
df['MaxCumReturn'] = df['CumReturn'].cummax()
# DD
df['DrawDown'] = (df['CumReturn'] / df['MaxCumReturn']) -1
# MDD
print(df['DrawDown'].min())
df.head(10)
```
